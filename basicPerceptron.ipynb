```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python


link='https://raw.githubusercontent.com/campusx-official/100-days-of-deep-learning/main/day3/placement.csv'

df = pd.read_csv(link)
df.head()
```

```python
plt.scatter(df['cgpa'], df['resume_score'], c=df['placed'], cmap='viridis', marker='o', edgecolors='k', s=100)
plt.title('Scatterplot of CGPA and Resume Score')
plt.xlabel('CGPA')
plt.ylabel('Resume Score')
plt.colorbar(label='Placed (1: Placed, 0: Not Placed)')
plt.grid(True)
plt.show()
```

```python
x=df.iloc[:,0:2]
y=df.iloc[:,-1]
```

```python
from sklearn.linear_model import Perceptron
p=Perceptron()
p.fit(x,y)
p.coef_,p.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x.values,y.values,clf=p,legend=2)
```
